## **Approach 1: Streak Counter**

This solution scans the array while **tracking the length of the current consecutive 1’s streak** and **keeping record of the longest streak found so far**.
It resets the counter whenever a `0` is encountered.

---

**Algorithm Description**

The method works as follows:

**1. Initialize Counters**

* Use two variables:

  ```python
  max_streak = 0       # Longest streak found so far
  current_streak = 0   # Length of the current consecutive 1’s streak
  ```

**2. Traverse the Array**

* For each element in `nums`:

  * **If the value is `1`**:

    * Increment `current_streak`:

      ```python
      current_streak += 1
      ```
    * Update `max_streak` if the current streak is longer:

      ```python
      max_streak = max(max_streak, current_streak)
      ```

  * **If the value is `0`**:

    * Reset `current_streak` to `0` since the streak is broken:

      ```python
      current_streak = 0
      ```

**3. Return the Result**

* After the loop, `max_streak` holds the maximum number of consecutive 1’s:

  ```python
  return max_streak
  ```

---

**Time and Space Complexity Analysis**

| Complexity | Explanation                                        |
| ---------- | -------------------------------------------------- |
| **Time**   | `O(n)` – Single pass through the array             |
| **Space**  | `O(1)` – Only two extra integer variables are used |

---

**Pros and Cons**

**✅ Pros:**

* **Optimal** in both time and space.
* Very readable — clear logic with minimal code.
* Handles arrays of any size without extra memory.
    
---

**Key Takeaway:**
This is the **standard and optimal** solution for finding the longest run of 1’s in a binary array — **simple, fast, and memory-efficient**.



````python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        max_streak = 0
        current_streak = 0

        for value in nums:
            if value == 1:
                current_streak += 1
                max_streak = max(max_streak, current_streak)
            else:
                current_streak = 0

        return max_streak
````
