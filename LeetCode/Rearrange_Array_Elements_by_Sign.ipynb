#### **Approach 1: Split Positive and Negative Numbers into Separate Lists and Interleave**

This approach separates the positive and negative numbers while preserving their original order, then rearranges them by alternating one positive and one negative number starting with a positive.

---

**Algorithm Explanation:**

1. **Create Two Separate Lists:**

   * Initialize two lists: one for positive integers `Positives`, and one for negative integers `Negatives`.

2. **Partition the Input Array:**

   * Iterate through each element in `nums`:

     * If the element is positive, append it to `Positives`.
     * If the element is negative, append it to `Negatives`.
   * This maintains the **original order** within each group.

3. **Reconstruct the Array Alternating Signs:**

   * Use a loop to place elements back into `nums`:

     * At even indices (`0, 2, 4, ...`), place elements from `Positives`.
     * At odd indices (`1, 3, 5, ...`), place elements from `Negatives`.

4. **Return the Modified Array:**

   * The result satisfies:

     * Alternating signs.
     * Starting with a positive number.
     * Original relative order within positives and negatives.

---

**Time Complexity:**

* Looping through `nums` once to separate elements: **O(n)**
* Looping again to merge into original array: **O(n)**
  **Total Time Complexity:** **O(2n)**

**Space Complexity:**

* Two additional lists of size `n/2`: **O(n)** for auxiliary space

---

```python
class Solution:
    def rearrangeArray(self, nums: List[int]) -> List[int]:
        Positives = []
        Negatives = []

        for Value in nums:
            if Value < 0:
                Negatives.append(Value)
            else:
                Positives.append(Value)

        for i in range(len(nums)//2):
            nums[2*i] = Positives[i]
            nums[2*i + 1] = Negatives[i]

        return nums
```
